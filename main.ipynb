# Settings

```python
# Settings 
images_number = 10
export_directory = "/content/drive/My Drive/Bible_Generation/" # Export Directory
```

# Main script




```python
df_passage = df_bible.query("chapter==1 & book_name=='Psalm'")

for index, row in df_passage.iterrows():

    export_file_path = f"{export_directory}{row['book_name']}/"

    # Check whether the specified path exists or not
    isExist = os.path.exists(export_file_path)
    if not isExist:

      # Create a new directory because it does not exist
      os.makedirs(export_file_path)
      print(f"The new directory is created: ", row['book_name'])

    for x in range(images_number):
      with autocast("cuda"):
        image = pipe(row["text"])["sample"][0]  
          
      image.save(f"{export_file_path}chaper{row['chapter']}_verse{row['verse']}_{x}.png")
    
```
